```python
X = pd.get_dummies(df.drop('species',axis=1),drop_first=True).astype(int)
X.head(10)
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_preds = model.predict(X_test)
```

```python
base_preds
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
```

```python
print(classification_report(y_test,base_preds))
```

```python
cm = confusion_matrix(y_test,base_preds, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
```

```python
model.feature_importances_
```

```python
X.columns
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance']).sort_values('Feature Importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(model)
```

```python
plt.figure(figsize=(12,8),dpi=200)
plot_tree(model);
```

```python
plt.figure(figsize=(12,8),dpi=200)
plot_tree(model, feature_names=X.columns, filled=True);
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=200)
    plot_tree(model, feature_names=X.columns, filled=True);
```

```python
report_model(model)
```

### For better performance lets try testing other hyperparameters

```python
# Setting a maximum depth for the tree or prunning your tree.
pruned_tree = DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
# Using the maximum leaf nnode hyperparameter.
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train,y_train)
```

```python
report_model(max_leaf_tree)
```

```python
# Using information gain or entropy tree.
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
