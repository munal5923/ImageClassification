## Logistic Regression Multi Class Problem.

```python
df = pd.read_csv('./DATA/iris.csv')
```

```python
df.head()
```

```python
sns.countplot(x='species',data=df)
```

```python
sns.pairplot(df,hue='species')
```

```python
num_df = df.drop('species', axis=1)
sns.heatmap(num_df.corr(),annot=True)
```

```python
X = df.drop('species', axis=1)
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
log_model = LogisticRegression(solver='saga',multi_class='ovr',max_iter=5000)
```

```python
penalty = ['l1','l2','elasticnet']
l1_ratio = np.linspace(0,1,20)
C = np.logspace(0,10,20)

param_grid = {'penalty':penalty, 'l1_ratio':l1_ratio,'C':C}
```

```python
grid_model = GridSearchCV(log_model,param_grid=param_grid)
```

```python
grid_model.fit(scaled_X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay,PrecisionRecallDisplay,RocCurveDisplay
```

```python
grid_model.best_params_
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred, labels=grid_model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=grid_model.classes_)
disp.plot()
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```
