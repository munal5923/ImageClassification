# Grid Search

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./DATA/Advertising.csv')
```

```python
df.head()
```

```python
## CREATE X and y
X = df.drop('sales',axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
# Our elastic net hyperparameters should be empty for now
base_elastic_net_model = ElasticNet()
```

```python
# Here we pass in the hyperparameters as it is in the elasticnet as a dictionary.
pram_grid = {'alpha':[0.1,1,5,10,50,100], 'l1_ratio':[.1,.5,.7,.95,.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,param_grid=pram_grid,scoring='neg_mean_squared_error',cv=5,
                         verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
# To get the actual results thst were not the best scoting parameters.
grid_model.cv_results_
```

```python
# passing it into a dataframe to make it more easier to read.
pd.DataFrame(grid_model.cv_results_)
```

```python
# Making your predictions
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```
